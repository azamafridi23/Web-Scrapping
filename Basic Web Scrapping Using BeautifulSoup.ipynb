```python
with open('home.html','r') as html_file:
    content = html_file.read()
    soup = BeautifulSoup(content,'lxml')
    
    course_cards = soup.find_all('div', class_='card')
    for card in course_cards:
        course_name = card.h5.text
        course_price =card.a.text.split() 
        print(course_name,'costs',course_price[-1])
```

```python
import requests # The requests module allows you to send HTTP requests using Python.
                #The HTTP request returns a Response Object with all the response data (content, encoding, status, etc).
                # Requess is used to request specific information from websites 
```

#### Link used to extract info = https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=data+science&txtLocation=

```python
html_text = requests.get('https://www.timesjobs.com/candidate/job-search.html?searchType=personalizedSearch&from=submit&txtKeywords=data+science&txtLocation=')
print(html_text) # 200 is standard code in web which means that request is successfull
```

```python
html_text = html_text.text # 
print(html_text)
```

```python
soup = BeautifulSoup(html_text,'lxml')
```

```python
job = soup.find('li',class_='clearfix job-bx wht-shd-bx')
print(job)
```

```python
company_name = job.find('h3',class_='joblist-comp-name').text.replace(' ','').replace('\r','').replace('\n','')
company_name
```

```python
skills = job.find('span',class_='srp-skills').text.replace('\n','').replace('\r','').strip()
skills
```

```python
published_date = soup.find('span',class_='sim-posted').text.replace('\n','').strip()
published_date
```

```python
link = job.h2.a
link
```

```python
link = link['href']
link
```

```python
print(f'Company name = {company_name}\nReq Skills: {skills}\nPublished Date = {published_date}\nLink = {link}')
```
